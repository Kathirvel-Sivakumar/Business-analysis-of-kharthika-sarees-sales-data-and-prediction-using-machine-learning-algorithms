```python
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
```

```python
num_pipeline = Pipeline(steps = [("imputer",SimpleImputer(strategy="mean")),
                                   ("sclaer",StandardScaler())])
```

```python
state_categories=['Tamilnadu','kerala','karnataka','andhra pradesh']
order_month_categories=['jan 21','feb 21','mar 21','apr 21','may 21','june 21','july 21','aug 21','sep 21','oct 21','nov 21','dec 21',
                        'jan 22','feb 22','mar 22','apr 22','may 22','june 22','july 22','aug 22','sep 22','oct 22','nov 22','dec 22']
customer_name_categories=['chennai silks','dev ganga silks','local customer','nesavu botique','psr silks','kumaran fabrics','nalli silks',
                          'ashoka handlooms','shanthi nikethan','thakadial sarees','mayoor millen sarees','raav silks','skm sarees',
                          'santhana brothers','keerthi creations','vineeth enterprises','muthu fabrics']
sarees_chudidhars_categories=['sarees','chudidhars']
type_categories=['pure silk','cotton','silk cotton','dye chudidhars','pure silk,silkcotton','cotton,silk cotton','dye sarees','pure silk',
                 'pure silk, silk cotton, cotton, dye sarees','cotton, dye sarees','pure silk, cotton','pure silk, dye sarees']
```

```python
cat_pipeline = Pipeline(steps = [("imputer",SimpleImputer(strategy='most_frequent')),
                                 ("ordinal encoder",OrdinalEncoder(categories=[state_categories,order_month_categories,customer_name_categories,
                                                      sarees_chudidhars_categories,type_categories])) ])
```

```python
from sklearn.compose import ColumnTransformer
```

```python
preprocessor =  ColumnTransformer(
    transformers=[("num_pipeline", num_pipeline, num_col),
                  ("cat_pipeline", cat_pipeline, cat_col)])
```

```python
X
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.20, random_state=42)
```

```python
print(X_train.shape) 
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
preprocessor.fit_transform(X_train)
preprocessor.transform(X_test)
```

```python
preprocessor.get_feature_names_out()
```

```python
X_train=pd.DataFrame(preprocessor.fit_transform(X_train),columns=preprocessor.get_feature_names_out())
X_test=pd.DataFrame(preprocessor.transform(X_test),columns=preprocessor.get_feature_names_out())
```

```python
from xgboost import XGBRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression,Lasso,Ridge,ElasticNet
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
```

```python
models={'LinearRegression':LinearRegression(),
        'Lasso':Lasso(),
        'Ridge':Ridge(),
        'Elasticnet':ElasticNet(),
        'Randomforest':RandomForestRegressor(),
         'xgboost':XGBRegressor()}
```

```python
trained_model_list=[]
model_list=[]
r2_list=[]
```

```python
list(models)
```

```python
for a in range(len(list(models))):
    model = list(models.values())[a]
    print(model)
```

```python
def evaluate_model(true,pred):
    r2=r2_score(true,pred)
    mae=mean_absolute_error(true,pred)
    mse=mean_squared_error(true,pred)

    return mae, mse,r2
```

```python
for i in range(len(list(models))):
    model=list(models.values())[i]
    
    model.fit(X_train,y_train)

    #make_prediction
    y_pred=model.predict(X_test)

    #this is for the validaiton
    MAE,MSE,R2=evaluate_model(y_test,y_pred)


    print("model training performance",model)
    print("MSE:", MSE)
    print("MAE:",MAE)
    print("R2 SCORE:",R2)

    r2_list.append(R2)

    print("="*40)
    print("\n")
```

```python
max(r2_list)
```

```python
pd.DataFrame(list(zip(models, r2_list)), columns=['Model Name', 'R2_Score']).sort_values(by=["R2_Score"],ascending=False)
```
