```python
df.columns=df.columns.str.lower()
```

```python
df=df.drop_duplicates()
```

```python
df
```

```python
Q1 = df['units'].quantile(0.25)
Q3 = df['units'].quantile(0.75)
IQR = Q3 - Q1
df = df[(df['units'] >= Q1 - 1.5 * IQR) & (df['units'] <= Q3 + 1.5 * IQR)]
```

```python
#which product has highest units of sales
plt.figure(figsize=(6,6))
sns.barplot(x='sarees/chudidhars',y='units',data=df)
plt.title('which product has highest units of sales')
plt.xlabel('sarees/chudidhars')
plt.ylabel('units')
plt.show()

plt.figure(figsize=(6,6))
sns.lineplot(x='sarees/chudidhars',y='units',data=df)
plt.title('which product has highest units of sales')
plt.xlabel('sarees/chudidhars')
plt.ylabel('units')
plt.show()

plt.figure(figsize=(6,6))
sns.scatterplot(x='sarees/chudidhars',y='units',data=df)
plt.title('which product has highest units of sales')
plt.xlabel('sarees/chudidhars')
plt.ylabel('units')
plt.show()

plt.figure()
plt.bar('sarees/chudidhars','units',data=df)
plt.title('which product has highest units of sales')
plt.xlabel('sarees/chudidhars')
plt.ylabel('units')
plt.show()

plt.figure(figsize=(6,6))
plt.hist('sarees/chudidhars',data=df)
plt.title('which product has highest units of sales')
plt.xlabel('sarees/chudidhars')
plt.ylabel('count')
plt.show()
```

```python
#which state had highest sales value
plt.figure(figsize=(6,6))
sns.scatterplot(x='state',y='total sales amount',data=df)
plt.title('which state has highest value of sales')
plt.xlabel('state')
plt.ylabel('total sales amount')
plt.show()

plt.figure()
plt.bar('state','total sales amount',data=df)
plt.title('which state has highest value of sales')
plt.xlabel('state')
plt.ylabel('total sales amount')
plt.show()

plt.figure(figsize=(6,6))
plt.hist('state',data=df)
plt.title('which state has highest value of sales')
plt.xlabel('state')
plt.ylabel('count')
plt.show()
```

```python
#which customer had highest sales value
plt.figure(figsize=(6,6))
sns.scatterplot(x='total sales amount',y='customer name',data=df)
plt.title('which customer had highest sales value')
plt.xlabel('total sales value')
plt.ylabel('customer name')
plt.show()
```

```python
#which type of sarees had highest unit of slaes
plt.figure(figsize=(6,6))
sns.barplot(x='units',y='type',data=df)
plt.title('which type of sarees had highest units of sales')
plt.xlabel('units')
plt.ylabel('type of sarees')
plt.show()

plt.figure(figsize=(50,10))
plt.bar('type','units',data=df)
plt.title('which type of sarees has highest units of sales')
plt.xlabel('type of sarees')
plt.ylabel('units')
plt.show()

plt.figure(figsize=(6,6))
sns.scatterplot(x='units',y='type',data=df)
plt.title('which type of sarees had highest units of sales')
plt.xlabel('units')
plt.ylabel('type of sarees')
plt.show()
```

```python
#which month had highest sales value
plt.figure(figsize=(6,6))
sns.scatterplot(x='total sales amount',y='order month',data=df)
plt.title('which month had highest sales value')
plt.xlabel('totall sales amount')
plt.ylabel('order month')
plt.show()

plt.figure(figsize=(6,6))
sns.barplot(x='total sales amount',y='order month',data=df)
plt.title('which month had highest sales value')
plt.xlabel('totall sales amount')
plt.ylabel('order month')
plt.show()
```
